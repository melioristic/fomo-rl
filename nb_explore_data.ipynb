```python
from fomo.io import read_benchmark_data
import matplotlib.pyplot as plt
import numpy as np
```

```python
ds_path = "/data/compoundx/benchmark-dataset/"

train, val, test = read_benchmark_data(ds_path=ds_path)
```

```python
# Explore test data

rand_ind = np.random.randint(10000)

# Plot random sample


Xd, Xs, Y = test
fig, (ax1, ax2, ax3) = plt.subplots(3,1)
ax1.plot(Xd[rand_ind, :, 0] )
ax2.bar(np.arange(36),  Xd[rand_ind, :, 1, 0] )
ax3.plot(Xd[rand_ind, :, 2] )

ax1.set_xlabel("Time in Months")
ax1.set_ylabel("Radiation")

ax2.set_xlabel("Time in Months")
ax2.set_ylabel("Precipitation")

ax3.set_xlabel("Time in Months")
ax3.set_ylabel("Temperature")

plt.tight_layout()
```

```python
fig, ax = plt.subplots(1,1)
ax.hist(Y, bins = 100)

ax.axvline(Y.mean(), color = "c", linestyle = "--", label = f"$\mu= {np.round(Y.mean(), 3)}$")
ax.axvline(np.median(Y), color = "k", linestyle = "--", label = f"median = {np.round(np.median(Y), 3)}")
ax.axvline(np.percentile(Y, 90), color = "r", linestyle = "--", label = f"90th percentile = {np.round(np.percentile(Y, 90), 3)}")


ax.legend()
plt.show()
```
